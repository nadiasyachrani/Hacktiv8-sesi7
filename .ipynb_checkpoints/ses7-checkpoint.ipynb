# BASIC VISUALIZATION
Visualisasi dasar pada Python adalah cara untuk merepresentasikan data dalam bentuk visual, seperti grafik, plot, diagram, atau gambar, dengan menggunakan library visualisasi seperti Matplotlib, Seaborn, Plotly, dan lain-lain. Visualisasi data dapat membantu pengguna memahami data yang kompleks atau besar, mengidentifikasi pola dan tren, membandingkan variabel, dan menemukan wawasan yang mungkin tersembunyi dalam data. Visualisasi dasar meliputi beberapa jenis plot dasar seperti scatter plot, line plot, bar plot, dan histogram. Dengan menggunakan library visualisasi di Python, pengguna dapat dengan mudah membuat visualisasi yang menarik dan informatif dengan kode Python yang relatif sederhana.

## Use Case 3
Compare top 5 countries that has most immigrations to Canada

```python
df_can.head()
```

```python
df_can.sort_values('Total', ascending=False, inplace=True)
df_can.head(5)
```

```python
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head(5)
df_top5 = df_top5[years].transpose()
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(15,10))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()

# sort values : mengurutkan nilai dari kolom total
# ascending=false jadinya descending
# diambil 5 teratas
# ditranspose agar tidak memanjang
```

## Area Plots
gabungan dari line plot dan bar plot


```python
df_top5.head()
```

```python
df_top5.index
```

```python
# Area pLOT
df_top5.plot(kind='area', stacked=False, figsize=(20,10))

#melihat trend satu persatu, negara

plt.title('Immigration trend of top 5 Countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

### Scripting Layer

```python
# Area pLOT - Scripting Layer
df_top5.plot(kind='area', alpha=0.25, stacked=False, figsize=(20,10))
# set transparancy pake alpha
# melihat trend satu persatu, negara
#stacked false agak burem

plt.title('Immigration trend of top 5 Countries')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')
plt.show()
```

```python
# Area pLOT
df_top5.plot(kind='area', stacked=True, figsize=(20,10))
plt.show()
#keseluruhan
```

### Artist Layer - Area Plot

```python
ax = df_top5.plot(kind='area', alpha=0.25, stacked=False, figsize=(20,10))
ax.set_title('Immigration Top 5 Countries')
ax.set_ylabel('Numbers of Immigrants')
ax.set_xlabel('years')
plt.show()
```

## Use Case : gunakan scripting layer
untuk membandingkan tren bottom 5 countries yang menjadi imigran

```python
df_can.tail(5) #5 terbawah
```

```python
df_least5 = df_can.tail(5)
df_least5 = df_least5[years].T

df_least5.plot(kind='area',alpha=0.35,stacked=False,figsize=(20,10))
plt.title('Immigration bottom 5 Countries')
plt.ylabel('Numbers of Immigrants')
plt.xlabel('years')
plt.show()
```

## Histogram


```python
df_can['2013'].head()
```

```python
# np.histogram bis ange-return 2 values
count, bin_edges = np.histogram(df_can['2013'])

print(count) #frekuensi
print(bin_edges) #bin ranges default=10 bin
```

```python
# plot menggunakan scripting layer
df_can['2013'].plot(kind='hist')
plt.title('histogram of immigration from 195 countries in 2013')
plt.ylabel('number of countries(frequency)')
plt.xlabel('number of immigrants')
plt.show()
```

```python
# np.histogram bis ange-return 2 values
count, bin_edges = np.histogram(df_can['2013'])

# plot menggunakan scripting layer
df_can['2013'].plot(kind='hist',figsize=(20,10))
plt.title('histogram of immigration from 195 countries in 2013')
plt.ylabel('number of countries(frequency)')
plt.xlabel('number of immigrants')
plt.show()
```

```python
df_can.loc[['Denmark','Norway','Sweden'], years]
```

```python
df_can.loc[['Denmark','Norway','Sweden'], years].T.plot(kind='hist',figsize=(10,6))
```

```python
df_T = df_can.loc[['Denmark','Norway','Sweden'],years].T
count,bin_edges = np.histogram(df_T,15)

# UNSTACK HISTOGRAM
df_T.plot(kind='hist',figsize=(10,6),bins=15,alpha=0.6,xticks=bin_edges,color=['coral','cadetblue','mediumseagreen'])
plt.title('histogram 3 negara')
plt.ylabel('number of countries(frequency)')
plt.xlabel('number of immigrants')
plt.show()
```

## Parameter Anotasi


```python
df_iceland.plot(kind='bar',figsize=(10,6),rot=90)
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')
plt.title('Iceland immigrants 1980-2013')
plt.annotate('',
            xy=(32,70),
            xytext=(28,20),
            xycoords='data', #menggunakan koordinat dari objek
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3',color='blue',lw=4))
# lw lebar
plt.annotate('2008-2011 financial crisis',
            xy=(28,30),
            rotation=72.5,
            va='bottom',
            ha='left')
```
